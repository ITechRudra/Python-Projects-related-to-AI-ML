```python
def dfs_iterative(graph, start):
    stack = [start]  # Step 1: Initialize stack
    visited = set()  # To track visited nodes

    while stack:  # Step 2: Run until stack is empty
        node = stack.pop()  # Step 5: Pop from stack

        if node not in visited:
            print(node, end=" ")  # Process the node
            visited.add(node)  # Step 4: Mark as visited

            # Step 3: Push all adjacent nodes (in reverse order for correct output)
            for neighbor in reversed(graph.get(node, [])):
                if neighbor not in visited:
                    stack.append(neighbor)

# Define the graph as an adjacency list
graph = {
    'A': ['B', 'C', 'D'],
    'B': ['E'],
    'C': ['F', 'G'],
    'D': ['H'],
    'E': ['I'],
    'F': ['J'],
    'G': [],
    'H': [],
    'I': [],
    'J': []
}

# Run DFS starting from 'A'
print("DFS Traversal (Iterative):")
dfs_iterative(graph, 'A')

#Written and Executed by:
#Name: Rudra Soni
#Enrollment No.: 24BT04D227

```

```python
from collections import deque

def BFS(graph, start, goal):
  queue = deque([start]) #Create a queue for BFS
  visited = set([start]) #Track visited nodes

  while queue:
    current_node = queue.popleft() #Deque a node from the queue
    print("Visiting: ",current_node)

    if current_node == goal: #Chech if the goal is reached
      print("Goal reached successfully, The node is: ",current_node)
      return True

    for neighbor in graph[current_node]: #Enqueue neighbors if not visited
      if neighbor not in visited:
        queue.append(neighbor)
        visited.add(neighbor)

graph1={
    'A' : ['B','C'],
    'B' : ['C'],
    'C' : ['D'],
    'D' : []
}
start_node = 'A'
goal_node = 'D'

print("BFS Traversal for graph 1: ")
BFS(graph1, start_node, goal_node)

graph2={
    'A': ['B', 'C', 'D'],
    'B': ['E'],
    'C': ['F', 'G'],
    'D': ['H'],
    'E': ['I'],
    'F': ['J'],
    'G': [],
    'H': [],
    'I': [],
    'J': []
}

start_node = 'A'
goal_node = 'J'

print("BFS Traversal for graph 2: ")
BFS(graph2, start_node, goal_node)

graph3={
    'H' : ['A'],
    'A' : ['B','D'],
    'B' : ['C','F'],
    'C' : ['G','E','H'],
    'G' : ['E','H'],
    'D' : ['F'],
    'F' : ['A'],
    'E' : ['F','B']
}

start_node = 'H'
goal_node = 'E'

print("BFS Traversal for graph 3: ")
BFS(graph3, start_node, goal_node)

graph4={
    'W': ['P', 'Q', 'R'],
    'P': ['S', 'U'],
    'Q': ['T'],
    'R': ['V'],
    'S': [],
    'U': [],
    'T': ['U'],
    'V': []
}

start_node = 'W'
goal_node = 'V'

print("BFS Traversal for graph 4: ")
BFS(graph4, start_node, goal_node)

graph5={
    1: [2, 5, 7],
    2: [3, 4],
    3: [],
    4: [3],
    5: [6],
    6: [],
    7: []
}

start_node = 1
goal_node = 6

print("BFS Traversal for graph 5: ")
BFS(graph5, start_node, goal_node)

#Written and Executed by:
#Name: Rudra Soni
#Enrollment No.: 24BT04D227
```

#Aim: Write a program to solve N-Queen Problem Solution

> AThis is a problem in which we have the value of n (No. of queens and the size of the board). Now the program will check all the possible solutions to fit all the queens on the board with the following criteria:

>1. No two queens can be in the same row.
>2. No two queens can be in the same column.
>3. No two queens can be in the same diagonal.  

```python
#Function to print Board
def Board(board):
  for row in board:
    print(" ".join(row))
  print("\n")

#Function to check row and column that if it's safe to place queen or not
def is_safe(board, row, col, n):
  for i in range(row):
    if board[i][col] == 'Q':
      return False

  #Function to check Upper-Left Diagonal that if it's safe to place queen or not
  for i, j in zip(range(row, -1, -1), range(col, n)):
    if board[i][j] == 'Q':
      return False
  #Function to check Upper-Right Diagonal that if it's safe to place queen or not
  for i, j in zip(range(row, -1, -1), range(col, n)):
    if board[i][j] == 'Q':
      return False

  return True #Safe to place queen


#Reccursive function to place queens row by row

def place_queen(board, row, n): #All queens placed sucessfully
  if row == n:
    Board(board)
    return

  for col in range(n):
    if is_safe(board, row, col, n): #Check if it's safe position
      board[row][col] = 'Q' #Place the queen
      place_queen(board, row + 1, n) #Move to the next row
      board[row][col] = '.' # BackTrack (remove queen)

#Main function to place queen on board with the given criteria

def n_queen(n):
  board = [['.'] * n for _ in range(n)]
  place_queen(board, 0, n)

#Example: Solve 4-Queen Problem
n_queen(4)

#Written and Executed by:
#Name: Rudra Soni
#Enrollment No.: 24BT04D227
```

#Aim: Write a program to create a Tic Tac Toe Game.

```python
#Write a program to create a Tic Tac Toe Game.

#First let's create the board for the game
def print_board(board):
    print("  1 | 2 | 3 ")
    print(" ---|---|---")
    for i, row in enumerate(board):
        print(f"{i + 1}  {row[0]} | {row[1]} | {row[2]}")
        if i < 2:
            print(" ---|---|---")

def check_winner(board):
    # Check rows
    for row in board:
        if row[0] == row[1] == row[2] and row[0] != ' ':
            return row[0]

    # Check columns
    for col in range(3):
        if board[0][col] == board[1][col] == board[2][col] and board[0][col] != ' ':
            return board[0][col]

    # Check diagonals
    if board[0][0] == board[1][1] == board[2][2] and board[0][0] != ' ':
        return board[0][0]
    if board[0][2] == board[1][1] == board[2][0] and board[0][2] != ' ':
        return board[0][2]

    return None

#Check if the board is full or not i.e. it is a new game or not
def is_full(board):
    for row in board:
        if ' ' in row:
            return False
    return True


#Defining players 'X' and 'O' and game playing function
def main():
    board = [[' ' for _ in range(3)] for _ in range(3)]
    players = ['X', 'O']
    turn = 0

    print("Welcome to Tic Tac Toe!")
    print_board(board)

    while True:
        print(f"Player {players[turn % 2]}'s turn!")
        try:
            row = int(input("Enter row (1-3): ")) - 1
            col = int(input("Enter column (1-3): ")) - 1

            if row not in range(3) or col not in range(3):
                print("Invalid input. Please enter numbers between 1 and 3.")
                continue
            if board[row][col] != ' ':
                print("This cell is already occupied. Choose another one.")
                continue

            board[row][col] = players[turn % 2]
            print_board(board)

            winner = check_winner(board)
            if winner:
                print(f"Player {winner} wins!")
                break

            if is_full(board):
                print("It's a draw!")
                break

            turn += 1

        except ValueError:
            print("Please enter a valid number.")

#Run the Game
if __name__ == "__main__":
    main()
```

#Practical 5
#Aim: Implement Water Jug Problem using state space search

```python
from collections import deque

def water_jug_problem(jug1_capacity, jug2_capacity, target):
    """
    Solves the Water Jug Problem using State Space Search with BFS.
    """
    visited = set()
    queue = deque()
    queue.append((0, 0, []))  # Initial state: (jug1_state, jug2_state, path)

    while queue:
        jug1, jug2, path = queue.popleft()

        # Goal state check
        if jug1 == target or jug2 == target:
            path.append((jug1, jug2))
            return path

        if (jug1, jug2) in visited:
            continue
        visited.add((jug1, jug2))

        # Generate possible next states (State Space Search)
        next_states = [
            (jug1_capacity, jug2),  # Fill Jug1
            (jug1, jug2_capacity),  # Fill Jug2
            (0, jug2),  # Empty Jug1
            (jug1, 0),  # Empty Jug2
            # Pour from Jug1 to Jug2
            (jug1 - min(jug1, jug2_capacity - jug2), jug2 + min(jug1, jug2_capacity - jug2)),
            # Pour from Jug2 to Jug1
            (jug1 + min(jug2, jug1_capacity - jug1), jug2 - min(jug2, jug1_capacity - jug1))
        ]

        for state in next_states:
            if state not in visited:
                queue.append((state[0], state[1], path + [(jug1, jug2)]))

    return "No solution possible"

# Taking user input
jug1_capacity = int(input("Enter capacity of Jug 1: "))
jug2_capacity = int(input("Enter capacity of Jug 2: "))
target = int(input("Enter target quantity: "))

# Solve the problem
solution = water_jug_problem(jug1_capacity, jug2_capacity, target)
print("Solution Path:", solution)

#Written and Executed by:
#Name: Rudra Soni
#Enrollment No.: 24BT04D227
#Div: 2 [Batch C1]
```

#Aim: Implement 8 Puzzle problem using A* Algorithm





```python
print("Goal State:")
print("___________________")
print("|  1  |  2  |  3  |")
print("___________________")
print("|  4  |  5  |  6  |")
print("___________________")
print("|  7  |  8  |  _  |")
print("___________________")
print("                   ")
print("                   ")
print("                   ")
print("Initial State:")
print("___________________")
print("|  1  |  5  |  7  |")
print("___________________")
print("|  3  |  8  |  _  |")
print("___________________")
print("|  6  |  2  |  4  |")
print("___________________")
```

```python
import heapq

class PuzzleNode:
    def __init__(self, state, parent=None, move=None, cost=0, heuristic=0):
        self.state = state
        self.parent = parent
        self.move = move
        self.cost = cost
        self.heuristic = heuristic
        self.total_cost = self.cost + self.heuristic

    def __lt__(self, other):
        return self.total_cost < other.total_cost


def heuristic(state, goal):
    """ Calculate the Manhattan Distance heuristic """
    distance = 0
    for i in range(3):
        for j in range(3):
            if state[i][j] != 0:
                x, y = divmod(goal.index(state[i][j]), 3)
                distance += abs(x - i) + abs(y - j)
    return distance


def get_neighbors(state):
    """ Generate all possible moves from the current state """
    moves = {
        'up': (-1, 0),
        'down': (1, 0),
        'left': (0, -1),
        'right': (0, 1)
    }
    neighbors = []

    x, y = [(i, row.index(0)) for i, row in enumerate(state) if 0 in row][0]

    for move, (dx, dy) in moves.items():
        new_x, new_y = x + dx, y + dy
        if 0 <= new_x < 3 and 0 <= new_y < 3:
            new_state = [row[:] for row in state]  # Deep copy
            new_state[x][y], new_state[new_x][new_y] = new_state[new_x][new_y], new_state[x][y]
            neighbors.append(PuzzleNode(new_state, move=move))  # Pass move correctly

    return neighbors


def a_star(start, goal):
    """ A* algorithm implementation for the 8-puzzle problem """
    goal_flat = [num for row in goal for num in row]
    open_list = []
    closed_set = set()

    start_node = PuzzleNode(start, cost=0, heuristic=heuristic(start, goal_flat))
    heapq.heappush(open_list, start_node)

    while open_list:
        current_node = heapq.heappop(open_list)

        if current_node.state == goal:
            path = []
            while current_node:
                path.append((current_node.move, current_node.state))
                current_node = current_node.parent
            return path[::-1]

        closed_set.add(tuple(tuple(row) for row in current_node.state))

        for neighbor_node in get_neighbors(current_node.state):
            neighbor_tuple = tuple(tuple(row) for row in neighbor_node.state)

            if neighbor_tuple in closed_set:
                continue

            neighbor_node.parent = current_node
            neighbor_node.cost = current_node.cost + 1
            neighbor_node.heuristic = heuristic(neighbor_node.state, goal_flat)
            neighbor_node.total_cost = neighbor_node.cost + neighbor_node.heuristic

            heapq.heappush(open_list, neighbor_node)

    return None


# Test case
start_state = [[1, 2, 3], [4, 0, 5], [6, 7, 8]]
goal_state = [[1, 2, 3], [4, 5, 6], [7, 8, 0]]

solution = a_star(start_state, goal_state)

if solution:
    for move, state in solution:
        if move:
            print(f"Move: {move}\nState:")
        for row in state:
            print(row)
        print()
else:
    print("No solution found.")



# Written and Executed by:
# Name: Rudra Soni
# Enroll No.: 24BT04D227
```

#Aim: Implement Travelling Salesmen solution using A Star Algorithm

```python
import heapq
```

```python
INF = 9999
```

```python
class State:
  def __init__(self, cost_so_far, current_city, visited_mask, path):
    self.g = cost_so_far
    self.city = current_city
    self.visited = visited_mask
    self.path = path

  def __lt__(self, other):
    return self.g < other.g

def tsp_a_star(graph):
    n = len(graph)
    all_visited = (1 << n) - 1

    def heuristic(curr_city, visited):
        min_edge = float('inf')
        for i in range(n):
            if not (visited >> i) & 1:
                min_edge = min(min_edge, graph[curr_city][i])
        return min_edge if min_edge != float('inf') else 0

    pq = []
    heapq.heappush(pq, (0, State(0, 0, 1, [0])))  # Start from city 0 (city 1 in real graph)

    while pq:
        f, curr = heapq.heappop(pq)

        if curr.visited == all_visited:
            total_cost = curr.g + graph[curr.city][0]
            return total_cost, curr.path + [0]

        for next_city in range(n):
            if not (curr.visited >> next_city) & 1 and graph[curr.city][next_city] != INF:
                new_visited = curr.visited | (1 << next_city)
                new_cost = curr.g + graph[curr.city][next_city]
                h = heuristic(next_city, new_visited)
                f_cost = new_cost + h
                heapq.heappush(pq, (f_cost, State(new_cost, next_city, new_visited, curr.path + [next_city])))

    return None
```

```python
# Distance matrix from your graph
graph = [
    [0, 12, 10, INF, INF, INF, 12],
    [12, 0, 8, 12, INF, INF, INF],
    [10, 8, 0, 11, 3, INF, 9],
    [INF, 12, 11, 0, 11, 10, INF],
    [INF, INF, 3, 11, 0, 6, 7],
    [INF, INF, INF, 10, 6, 0, 9],
    [12, INF, 9, INF, 7, 9, 0]
]

cost, path = tsp_a_star(graph)
```

```python
# Output
print("Minimum Cost:", cost)
print("Path:", ' -> '.join(str(p+1) for p in path))  # +1 because our list is 0-indexed
```
